### *Visualizations*

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Distribution of different features represented using boxplot

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='SL',data=data)
plt.title('Distribution of sepal length for each classification')

plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='SW',data=data)
plt.title('Distribution of sepal width for each classification')

plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='PL',data=data)
plt.title('Distribution of petal length for each classification')

plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='PW',data=data)
plt.title('Distribution of petal width for each classification')
```

##### Relation between sepal length and width and petal length and width presented using scatterplot

```python
plt.figure(figsize=(10,10))

plt.subplot(2,1,1)
sns.scatterplot(x='SL',y='SW',hue='Classification',data=data)
plt.title('Relation between sepal length and sepal width of all species ')

plt.subplot(2,1,2)
sns.scatterplot(x='PL',y='PW',hue='Classification',data=data)
plt.title('Relation between petal length and petal width of all species')
```

CONCLUSIONS
* Iris-setosa has shorter sepal length and larger sepal width.
* Iris-virginica has comparitively longer sepal length and shorter sepal width.
* Iris-setosa has shorter petal length and width while Iris-virginica has larger petal length and width.

##### Correlation between different features in the data

```python
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=45)
```

CONCLUSIONS
* Relation between petal width and petal length is high.
* Relation between petal width and sepal length is high.
